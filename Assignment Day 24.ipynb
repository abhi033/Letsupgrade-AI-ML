```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

Import given dataset (bank loan model dataset)

```python
df = pd.read_excel('Bank_Personal_Loan_Modelling.xlsx',sheet_name=1)
df.head()
```

```python
df.columns
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.drop(['ID'],axis=1,inplace=True)
```

```python
df.info()
```

TO get random forest of the dataset

```python
randomForestClassifier = RandomForestClassifier(n_estimators=1000,max_features=12,oob_score=True)
```

```python
randomForestClassifier.fit(df.drop(['Personal Loan'],axis=1),df['Personal Loan'])
```

To get dicision tree of the dataset 

```python
decisionTreeClassifier = DecisionTreeClassifier(max_depth=3)
```

```python
decisionTreeClassifier.fit(df[['Family','Income','Education']],df['Personal Loan'])
```

```python
plt.rcParams["figure.figsize"] = (300,100)
```

```python
Plotting tree of the decision tree classifier
```

```python
tree.plot_tree(decisionTreeClassifier,filled = True)
```
